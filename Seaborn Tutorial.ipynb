```python
import seaborn as sns
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

# Distribution plot
* distplot
* jointplot
* pairplot
* kdeplot

#### distploy
Shows the distribution of a Univarient set of Observations

```python
#KDE => kernel density estimation
sns.distplot(tips['total_bill'],kde=False,bins=50)
```

#### jointplot
Plot 2 data points (Bi-Varient)
* scatter diagram
* regression
* kde
* hex

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='scatter')
```

#### pairplot
Visualization of relationshipt of all numerical values

```python
sns.pairplot(tips,hue='sex')
```

### Kdeplot

```python
sns.kdeplot(tips['total_bill'])
```

# Categorical Data Plots
* catplot
* boxplot
* voilinplot
* stripplot
* swarmplot
* barplot
* countplot

#### barplot and countplot

```python
import numpy as np
```

```python
sns.barplot(x='sex',y='total_bill',data=tips,hue='time',estimator=np.std)
```

```python
# This shows whether our data sets is balanced or not.
sns.countplot(x='sex',data=tips)
```

#### boxplot
To seet the outlier in our data sets.

```python
sns.boxplot(x='day',y='total_bill',data=tips, hue='smoker')
```

```python
sns.violinplot(x='day',y='total_bill',data=tips, hue='sex',split=True)
```

### stripplot and swarmplot

```python
sns.stripplot(x='day',y='total_bill',data=tips,hue='sex',dodge=True)
```

```python
sns.swarmplot(x='day',y='total_bill',data=tips,hue='sex',dodge=True)
```

```python
# Universal plot
sns.catplot(x='day',y='total_bill',data=tips,hue='sex',kind='bar')
```

# Grids

```python
iris = sns.load_dataset('iris')
```

```python
iris.head()
```

## PairGrid

```python
import matplotlib.pyplot as plt
```

```python
g = sns.PairGrid(iris)
g.map(plt.scatter)
```

```python
g = sns.PairGrid(iris)
g.map_diag(plt.hist)
g.map_upper(plt.scatter)
g.map_lower(sns.kdeplot)
```

```python
g= sns.FacetGrid(tips,col='time',row='smoker',hue='sex')
# g.map(plt.hist,'total_bill')
g.map(plt.scatter,'total_bill','tip').add_legend()
```

# JointGrid

```python
g= sns.JointGrid(x='total_bill',y='tip',data=tips)
g.plot(sns.regplot,sns.distplot)
```
